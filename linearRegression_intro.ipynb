```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np

# Set the random seed for reproducibility

np.random.seed(42)

# Generate a random regression dataset

X, y = make_regression(n_samples=500, n_features=1, noise=20)

# Plot the generated dataset

plt.scatter(X, y)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Random Regression Dataset')
plt.show()
```

```python
from  sklearn.linear_model import LinearRegression

# Create a linear regression model

model = LinearRegression()

# Fit the model to the dataset

model.fit(X, y)

# Print the intercept and coefficient of the linear regression model

intercept = model.intercept_

coefficient = model.coef_[0]

# Generate a range of values for the feature

x_values = np.linspace(np.min(X), np.max(X))

# Calculate the predicted values for the feature

y_values = intercept + coefficient * x_values

# Plot the generated dataset and the linear regression model

plt.scatter(X, y)
plt.plot(x_values, y_values, color='red')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Random Regression Dataset and Linear Regression Model')
```

### Regressão Linear: Interpretação da Fórmula e Visualização

Neste exemplo, utilizamos o modelo de **Regressão Linear** para ajustar uma reta aos dados e prever o valor da variável alvo (`y`) com base na variável de entrada (`X`). A **regressão linear simples** é descrita pela fórmula:

$
y = b_0 + b_1 \cdot X
$

onde:
- **\( y \)**: Valor previsto para a variável dependente (ou alvo).
- **\( b_0 \)**: Intercepto da reta, o ponto em que a reta cruza o eixo y.
- **\( b_1 \)**: Coeficiente da reta (ou inclinação), que representa a variação de \( y \) para cada unidade de \( X \).

# Regressão Linear

A **regressão linear** é uma técnica de modelagem estatística usada para prever o valor de uma variável dependente \( y \) com base em uma ou mais variáveis independentes \( x \). A forma mais simples é o **modelo de regressão linear simples**, expresso pela equação:

$$
\text{reg} = m \cdot x + b
$$

onde:
- \( m \) é o coeficiente angular (a inclinação da reta),
- \( b \) é o coeficiente linear (intercepto da reta com o eixo \( y \)),
- \( x \) é a variável independente,
- \( \text{reg} \) é o valor predito pela regressão.

## 1. Função de Custo (Erro Quadrático Médio)

A função de custo mede o quanto os valores previstos pelo modelo estão distantes dos valores reais, sendo calculada pela soma dos quadrados das diferenças entre o valor real \( y_i \) e o valor predito \( \text{reg}_i \):

$$
\text{Custo} = \sum_{i=1}^{m} (\text{reg}_i - y_i)^2
$$

onde:
- \( m \) representa o número de observações,
- \( \text{reg}_i \) é o valor predito pelo modelo para a i-ésima observação,
- \( y_i \) é o valor real correspondente.

A função de custo visa minimizar essa diferença, ajustando os parâmetros \( m \) e \( b \) da equação da reta de regressão.

## 2. Gradiente Descendente

O **gradiente descendente** é um algoritmo que encontra os valores ideais dos coeficientes \( m \) e \( b \) que minimizam a função de custo. A cada iteração, os valores são ajustados pelas seguintes equações:

$$
m \leftarrow m - \alpha \cdot \frac{\partial \text{Custo}}{\partial m}
$$

$$
b \leftarrow b - \alpha \cdot \frac{\partial \text{Custo}}{\partial b}
$$

onde:
- \( \alpha \) é a **taxa de aprendizado**, controlando o tamanho do passo,
- \( \frac{\partial \text{Custo}}{\partial m} \) e \( \frac{\partial \text{Custo}}{\partial b} \) são as derivadas parciais da função de custo em relação a \( m \) e \( b \).

# Derivada da Função de Custo

## 1. Erro da Regressão

O erro \( \text{erro}_i \) para uma observação \( i \) pode ser expresso como a diferença entre o valor predito pelo modelo e o valor real \( y_i \):

$$
\text{erro}_i = m \cdot x_i + b - y_i
$$

onde:
- \( m \) é a inclinação da reta,
- \( x_i \) é o valor da variável independente na i-ésima observação,
- \( b \) é o intercepto,
- \( y_i \) é o valor real da variável dependente.

## 2. Função de Custo

A função de custo é a soma dos erros ao quadrado para todas as observações:

$$
\text{Custo} = \sum_{i=1}^{m} (\text{reg}_i - y_i)^2 = \sum_{i=1}^{m} (\text{erro}_i)^2
$$

Essa função mede o quão longe a reta de regressão está dos pontos reais.

## 3. Derivadas Parciais da Função de Custo

Para otimizar os valores de \( m \) e \( b \), utilizamos o **gradiente descendente**, que exige o cálculo das derivadas parciais da função de custo em relação a \( m \) e \( b \).

### Derivada Parcial em Relação a \( m \):

$$
\frac{\partial \text{Custo}}{\partial m} = \sum_{i=1}^{m} 2 \cdot \text{erro}_i \cdot x_i
$$

Aqui, multiplicamos o erro \( \text{erro}_i \) pelo valor correspondente de \( x_i \) e somamos para todas as observações.

### Derivada Parcial em Relação a \( b \):

$$
\frac{\partial \text{Custo}}{\partial b} = \sum_{i=1}^{m} 2 \cdot \text{erro}_i
$$

Neste caso, somamos apenas os erros para calcular o impacto de \( b \) na função de custo.
